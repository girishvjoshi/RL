```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

# File Name for saving the Results to file
file_name = 'hw4_PGgridWorldmlp_easyon'

with open(file_name,'rb') as file:
    PG_avgReward_easyon = pickle.load(file)
    
file_name = 'hw2_qlearning_test1'

with open(file_name,'rb') as file:
    QLearn_avgReward_easyon = pickle.load(file)
    
file_name = 'hw2_sarsa_test1'

with open(file_name,'rb') as file:
    SARSA_avgReward_easyon = pickle.load(file)
    
plt.figure(1)
plt.plot(PG_avgReward_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('Policy Gradient: Deterministic Gid World')
plt.show()

plt.figure(2)
plt.plot(QLearn_avgReward_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('Q-Learning: Deterministic Gid World')
plt.show()

plt.figure(3)
plt.plot(SARSA_avgReward_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('SARSA: Deterministic Gid World')
plt.show()
```
