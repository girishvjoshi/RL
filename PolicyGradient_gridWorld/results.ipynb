```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

# File Name for saving the Results to file
file_name = 'hw4_PGgridWorldmlp_easyon'

with open(file_name,'rb') as file:
    PG_avgRewardmlp_easyon = pickle.load(file)
    
file_name = 'hw2_qlearning_test1'

with open(file_name,'rb') as file:
    QLearn_avgReward_easyon = pickle.load(file)
    
file_name = 'hw2_sarsa_test1'

with open(file_name,'rb') as file:
    SARSA_avgReward_easyon = pickle.load(file)
    
plt.figure(1)
plt.plot(PG_avgRewardmlp_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('Policy Gradient: Deterministic Grid World')
plt.show()

plt.figure(2)
plt.plot(QLearn_avgReward_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('Q-Learning: Deterministic Grid World')
plt.show()

plt.figure(3)
plt.plot(SARSA_avgReward_easyon)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('SARSA: Deterministic Grid World')
plt.show()
```

The above are the comparative results for policy Gradient Qlearning and SARSA.

Using a multilayer perceptron as

action = softmax(relu(w2^T relu(w1^TX)))

The performance is similar to Q-Learning or SARSA

```python
# File Name for saving the Results to file
file_name = 'hw4_PGgridWorldmlp_easyoff'

with open(file_name,'rb') as file:
    avgRewardmlp_easyoff = pickle.load(file)
    
plt.plot(avgRewardmlp_easyoff)
plt.grid(axis='both')
plt.xlabel('Epochs')
plt.ylabel('Average Reward')
plt.title('Policy Gradient: Stochastic Grid World')
plt.show()
```
