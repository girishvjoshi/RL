```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

# File Name for saving the Results to file
file_name = 'hw2_qlearning_easyon'

with open(file_name,'rb') as file:
    avgReward = pickle.load(file)

# Printing Average Reward for every N epochs  
plt.plot(avgReward)
plt.xlabel('Epochs')
plt.ylabel('Avg Reward')
plt.title('Q-Learning - Deterministic GridWorld')
plt.grid(axis='both')
plt.show()
```
